```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\sreen\\Downloads\\PS_dataset.csv')
qscore=df['quality'].unique()
cp=df['quality'].value_counts()
c=cp.values.tolist()
plt.bar(qscore,c, color='violet')
plt.xlabel('Quality_score')
plt.ylabel('Count_of_wines')
plt.title('Bar_plot')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\sreen\\Downloads\\PS_dataset.csv')
plt.scatter(df['pH'], df['alcohol'])
plt.title('Scatter_Plot_of_Alcohol_vs_pH')
plt.xlabel('pH_value')
plt.ylabel('Alcohol_content (%)')
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('C:\\Users\\sreen\\Downloads\\PS_dataset.csv') 
plt.hist(df['alcohol'], bins=40, edgecolor='blue')
plt.title('Histogram of Alcohol Content')
plt.xlabel('Alcohol Content (%)')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('C:\\Users\\sreen\\Downloads\\PS_dataset.csv')  
plt.figure(figsize=(12,11))
df.boxplot(column='residual sugar', by='quality', grid=False)
plt.title('Box_Plot_of_Residual_Sugar_by_Quality_Score')
plt.suptitle('')
plt.xlabel('Quality_score')
plt.ylabel('Residual_sugar')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('C:\\Users\\sreen\\Downloads\\PS_dataset.csv') 
correlation_matrix = df.corr()
plt.figure(figsize=(14,9))
sns.heatmap(correlation_matrix, annot=True, fmt='.4f', cmap='magma', square=True, cbar_kws={"shrink": 1.1})
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('C:\\Users\\sreen\\Downloads\\PS_dataset.csv') 
plt.figure(figsize=(15,8))
plt.plot(df['free sulfur dioxide'], df['total sulfur dioxide'], marker='o', linestyle='--', color='purple')
plt.title('Trend_of_Total_Sulfur_Dioxide_vs_Free_Sulfur_Dioxide')
plt.xlabel('Free_sulfur_ioxide (mg/L)')
plt.ylabel('Total_sulfur_dioxide (mg/L)')
plt.grid()
plt.show()
```
